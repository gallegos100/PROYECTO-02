```python
def poligono(L):
    P1x=L/2*-1 
    P1y=-1*L*pow(3,0.5)*0.5
    P2x=L/2 
    P2y=-1*L*pow(3,0.5)*0.5
    P3x=L 
    P3y=0
    P4x=L/2 
    P4y=L*pow(3,0.5)*0.5
    P5x=-1*L/2 
    P5y=L*pow(3,0.5)*0.5
    P6x=-1*L 
    P6y=0
    x=[P1x,P2x,P3x,P4x,P5x,P6x,P1x]
    y=[P1y,P2y,P3y,P4y,P5y,P6y,P1y]
    plt.plot(x,y, ":",color="b")
    return plt.show()
```

```python
poligono(10)
```

MAPA DE CALOR

```python
# unstack permite disponer la agrupación en forma tabular (para el heatmap)
ratings = data_autos.groupby(['Anyo', 'Pais_Origen'])['Millas_Por_Galon'].mean().unstack()

ratings
```

```python
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)

text_colors = ('black', 'white')
im = ax.imshow(ratings, cmap='Reds')  # mapa de calor
cbar = fig.colorbar(im, ax=ax, label='Millas_Por_Galon')  # leyenda
cbar.outline.set_visible(False)


x = ratings.columns
y = ratings.index

# Formato de los ejes
ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation=90)
ax.set_yticks(range(len(y)))
ax.set_yticklabels(y)
ax.invert_yaxis()


#Presentacion
fig.tight_layout()
```
